# Generate training and testing data
## Case 1: Training Data is linerally separable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sci
```

```python
np.random.seed(42)
# Make sure there are 30 points
num_points = 30

# x for case 1 and case 2
x_cases = np.random.randint(low = 0, high = 25, size = num_points, dtype = int)
print(x_cases)

x_above = x_cases[:num_points // 2]
x_below = x_cases[num_points // 2:]

# Line that points will be above and below
x = np.linspace(-10, 30, num = 50)
m, b = 3, 2
y_line = m*x + b

plt.plot(x, y_line)

# Generate y-values for linearally separable case
y_above = []
y_below = []

for i in range(len(x_above)):
    y_high = np.random.randint(low=y_line[i] + 2, high=50, size=1, dtype=int)
    y_above.append(y_high)

for i in range(len(x_below)):
    y_low = np.random.randint(low=-50, high=y_line[i] - 2, size=1, dtype=int)
    y_below.append(y_low)

print('Y_values above line:', y_above)
print()
print('Y-values below line:', y_below)


plt.scatter(x_above, y_above, color = 'blue')
plt.scatter(x_above, y_below, color = 'red')
```
